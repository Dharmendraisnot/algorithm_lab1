```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
x=print(func4([1,2,3,4],[2,4,6,8]))
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    if not arr:
        return None, None

    max_val = min_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
        elif num < min_val:
            min_val = num

    return max_val, min_val
my_arr = [3, 5, 1, 9, 2, 8]
largest, smallest = max_min(my_arr)
print(f"Max: {largest}, Min: {smallest}")
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count
arr = [12, 1234, 45, 67, 1]
evens, odds = Counting_EvenOdd(arr)

print(f"Even numbers: {evens}")
print(f"Odd numbers: {odds}")
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    for index in range(len(arr)):
        if arr[index] == target:
            return index
    return -1
arr = [10, 23, 45, 70, 11, 15]
target_val = 70

result = linear_Search(arr, target_val)
print(f"Element found at index: {result}")
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
arr = [1, 3, 5, 7, 9]
target = 7
index = linear_Search(arr, target)
if index != -1:
    print(f"Target {target} found at index: {index}")
else:
    print(f"Target {target} not found in array.")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid

        elif arr[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9, 15, 16, 14, 45]
arr.sort()
target = 14
index = Binary_Search(arr, target)
print(f"Sorted Array: {arr}")
print(f"Target {target} found at index: {index}")
```
